Método BOX MULLER: Crear una VA que distribuya como NORMAL (0,1) a partir de dos uniformes (0, 1). ¿Son independientes? Sacar correlación

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
u1 = stats.uniform(0, 1)
u2 = stats.uniform(0, 1)

x = np.sqrt(-2 * np.log(u1.rvs(1000))) * np.cos(2 * np.pi * u2.rvs(1000))
y = np.sqrt(-2 * np.log(u1.rvs(1000))) * np.sin(2 * np.pi * u2.rvs(1000))

# Calcular la correlación
correlacion = np.corrcoef(x, y)[0, 1]
print(correlacion)

# Crear gráfico de dispersión
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, alpha=0.5)
plt.title(f'Gráfico de Dispersión\nCorrelación: {correlacion:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Crear la distribución normal
n = stats.norm(0, 1)

# Calcular la curva de la distribución normal
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = n.pdf(x_vals)

# Crear histogramas normalizados
plt.subplot(1, 2, 2)
plt.hist(x, bins=20, density=True, alpha=0.6, color='g', label='x')
plt.hist(y, bins=20, density=True, alpha=0.6, color='b', label='y')

# Superponer la curva de la distribución normal
plt.plot(x_vals, pdf_vals, 'k', linewidth=2, label='Distribución Normal')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
